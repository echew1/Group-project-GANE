```python
import matplotlib.pyplot as plt
import numpy as np

# Define substrate concentration values
s = np.linspace(0, 10, 100)

# Define enzyme activity function without inhibitor
def enzyme_activity_no_inhibitor(s, vmax, km):
    return (vmax * s) / (km + s)

# Define enzyme activity function with inhibitor
def enzyme_activity_with_inhibitor(s, vmax, km, ki):
    return (vmax * s) / (km + s * (1 + ki / km))

# Set enzyme kinetic parameters
vmax = 1.0
km = 0.5
ki = 0.2

# Calculate enzyme activity values without inhibitor
activity_no_inhibitor = enzyme_activity_no_inhibitor(s, vmax, km)

# Calculate enzyme activity values with inhibitor
activity_with_inhibitor = enzyme_activity_with_inhibitor(s, vmax, km, ki)

# Generate the competitive inhibitors graph
fig, ax = plt.subplots()
ax.plot(s, activity_no_inhibitor, label='No Inhibitor')
ax.plot(s, activity_with_inhibitor, label='With Inhibitor')
ax.set_xlabel('Substrate Concentration')
ax.set_ylabel('Enzyme Activity')
ax.set_title('Competitive Inhibitors Graph')
ax.legend()

plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the function to fit
def sigmoid(x, L, x0, k):
    y = L / (1 + np.exp(-k*(x-x0)))
    return y

# Generate some example data
xdata = np.array([0.1, 0.3, 0.5, 0.7, 0.9])
ydata = np.array([0.1, 0.4, 0.7, 0.9, 1.0])

# Fit the function to the data
popt, pcov = curve_fit(sigmoid, xdata, ydata)

# Generate points to plot the curve
x = np.linspace(0, 1, 100)
y = sigmoid(x, *popt)

# Plot the data points and the curve
plt.plot(xdata, ydata, 'o', label='data')
plt.plot(x,y, label='fit')
plt.xlabel('Dose')
plt.ylabel('Response')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the dose-response equation for a sigmoidal curve
def sigmoidal_dose_response(efficacy, potency, hill_slope, dose):
    return efficacy / (1 + (potency / dose) ** hill_slope)

# Define the parameters for paracetamol
paracetamol_efficacy = 1.0
paracetamol_potency = 10.0
paracetamol_hill_slope = 2.0

# Define the parameters for ibuprofen
ibuprofen_efficacy = 1.0
ibuprofen_potency = 20.0
ibuprofen_hill_slope = 1.0

# Generate the dose-response curves for paracetamol and ibuprofen
dose = np.linspace(0, 100, 100)
paracetamol_response = sigmoidal_dose_response(paracetamol_efficacy, paracetamol_potency, paracetamol_hill_slope, dose)
ibuprofen_response = sigmoidal_dose_response(ibuprofen_efficacy, ibuprofen_potency, ibuprofen_hill_slope, dose)

# Generate the dose-response curve graph
fig, ax = plt.subplots()
ax.plot(dose, paracetamol_response, label='Paracetamol')
ax.plot(dose, ibuprofen_response, label='Ibuprofen')
ax.set_xlabel('Dose')
ax.set_ylabel('Response')
ax.set_title('Dose-Response Curve')
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the pharmacokinetic equation for a one-compartment model
def one_compartment_model(dose, volume, absorption_rate, elimination_rate, time):
    ka = absorption_rate / volume
    ke = elimination_rate / volume
    fraction_absorbed = ka / (ka + ke)
    c0 = dose / volume
    cp = c0 * fraction_absorbed * np.exp(-ka * time) * (1-np.exp(-(ka+ke)*time)) / (ka + ke)
    cp += c0 * (1 - fraction_absorbed) * np.exp(-ke * time)
    return cp

# Define the parameters for paracetamol
paracetamol_dose = 500.0
paracetamol_volume = 50.0
paracetamol_absorption_rate = 2.0
paracetamol_elimination_rate = 1.0

# Define the parameters for ibuprofen
ibuprofen_dose = 500.0
ibuprofen_volume = 100.0
ibuprofen_absorption_rate = 1.0
ibuprofen_elimination_rate = 2.0

# Generate the concentration-time curves for paracetamol and ibuprofen
time = np.linspace(0, 450, 1000)
paracetamol_concentration = one_compartment_model(paracetamol_dose, paracetamol_volume, 
    paracetamol_absorption_rate, paracetamol_elimination_rate, time)
ibuprofen_concentration = one_compartment_model(ibuprofen_dose, ibuprofen_volume, 
    ibuprofen_absorption_rate, ibuprofen_elimination_rate, time)

# Generate the concentration-time curve graph
fig, ax = plt.subplots()
ax.plot(time, paracetamol_concentration, label='Paracetamol')
ax.plot(time, ibuprofen_concentration, label='Ibuprofen')
ax.set_xlabel('Time')
ax.set_ylabel('Concentration')
ax.set_title('Concentration-Time Curve')
ax.legend()

plt.show()
plt.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define pharmacokinetic parameters for paracetamol
paracetamol_CL = 50.0 # Clearance (mL/min)
paracetamol_Vd = 5000.0 # Volume of distribution (mL)
paracetamol_F = 1.0 # Bioavailability
paracetamol_Ka = 0.0 # Absorption rate constant

# Define pharmacokinetic parameters for ibuprofen
ibuprofen_CL = 30.0 # Clearance (mL/min)
ibuprofen_Vd = 3000.0 # Volume of distribution (mL)
ibuprofen_F = 1.0 # Bioavailability
ibuprofen_Ka = 0.0 # Absorption rate constant

# Define dosing parameters
paracetamol_dose = 1000.0 # Paracetamol dose (mg)
ibuprofen_dose = 400.0 # Ibuprofen dose (mg)
dosing_time = 0.0 # Dosing time (hours)

# Define time parameters
time = np.linspace(0, 12, 1000) # Time (hours)

# Calculate concentration-time profiles for paracetamol and ibuprofen
paracetamol_conc = paracetamol_dose * paracetamol_F / (paracetamol_Vd * np.exp(-paracetamol_CL / paracetamol_Vd * (time - dosing_time)))
ibuprofen_conc = ibuprofen_dose * ibuprofen_F / (ibuprofen_Vd * np.exp(-ibuprofen_CL / ibuprofen_Vd * (time - dosing_time)))

# Plot concentration-time profiles for paracetamol and ibuprofen
plt.plot(time, paracetamol_conc, label='Paracetamol')
plt.plot(time, ibuprofen_conc, label='Ibuprofen')
plt.xlabel('Time (hours)')
plt.ylabel('Concentration (mg/mL)')
plt.title('Concentration versus Time Profiles')
plt.legend()
plt.show()
```
